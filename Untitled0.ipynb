```python
import numpy as np
import polars as pl
import seaborn as pltsns # псевдоним чтобы не конфликтовать
from catboost import CatBoostClassifier, Pool
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
file_path = "likes.parquet"
data = kagglehub.dataset_load(
    KaggleDatasetAdapter.POLARS,
    "thekabeton/ysda-recsys-2026-yambda-dataset/versions/3",
    file_path
).collect()

test_users = kagglehub.dataset_load(
    KaggleDatasetAdapter.POLARS,
    "thekabeton/ysda-recsys-2026-yambda-dataset/versions/3",
    "test_users.csv"
).collect()

artists = kagglehub.dataset_load(
    KaggleDatasetAdapter.POLARS,
    "thekabeton/ysda-recsys-2026-yambda-dataset/versions/3",
    "artist_item_mapping_small.parquet"
).collect()
```

```python
# --- 2. ОПТИМИЗАЦИЯ ОЗУ И ВРЕМЕННОЙ СПЛИТ (Совет Алексея и Ильгиза) ---
DAY_IN_SEC = 24 * 60 * 60
max_ts = data['timestamp'].max()

# Берем только последние 21 день. Это ОЧЕНЬ сильно разгрузит ОЗУ и повысит актуальность
recent_data = data.filter(pl.col("timestamp") >= (max_ts - 21 * DAY_IN_SEC))

# Бьем на train и val по ВРЕМЕНИ, а не по длине.
# Val: последний день. Train: предпоследний день. History: все что до них.
val_start = max_ts - 1 * DAY_IN_SEC
train_start = max_ts - 2 * DAY_IN_SEC

test_uids = test_users.select('uid').unique()
```

```python
# Формируем историю и таргеты.
# Чтобы не упасть по ОЗУ, учиться будем только на пользователях из test_users
hist_train = recent_data.filter(pl.col("timestamp") < train_start)
target_train = recent_data.filter((pl.col("timestamp") >= train_start) & (pl.col("timestamp") < val_start))
target_train = target_train.join(test_uids, on='uid', how='inner')

hist_val = recent_data.filter(pl.col("timestamp") < val_start)
target_val = recent_data.filter(pl.col("timestamp") >= val_start)
target_val = target_val.join(test_uids, on='uid', how='inner')

hist_test = recent_data # Вся история для финального предикта
```

```python
def get_recent_candidates(history_df: pl.DataFrame, top_n: int = 300) -> pl.DataFrame:
    # Ищем популярное именно за недавнее время
    return (
        history_df.group_by("item_id").len()
        .sort("len", descending=True)
        .head(top_n)
        .select("item_id")
    )
```

```python
def create_dataset_with_hard_negatives(history: pl.DataFrame, target: pl.DataFrame, n_negatives: int = 10) -> pl.DataFrame:
    # 1. Генерируем кандидатов (топ 300)
    candidates = get_recent_candidates(history, top_n=300)

    # 2. Позитивы: то, что юзер реально лайкнул из таргета
    pos = target.select(["uid", "item_id"]).unique().with_columns(pl.lit(1).alias("target"))

    # 3. База для негативов: скрещиваем юзеров из таргета со всеми кандидатами
    target_users = target.select("uid").unique()
    all_cands = target_users.join(candidates, how="cross")

    # 4. Выкидываем из кандидатов то, что юзер лайкнул (чтобы не было ложных негативов)
    neg = all_cands.join(pos, on=["uid", "item_id"], how="anti")

    # 5. Сэмплируем негативы (берем примерно n_negatives на каждый позитив)
    neg_sampled = neg.sample(n=pos.height * n_negatives, with_replacement=True, seed=42).with_columns(pl.lit(0).alias("target"))

    # Объединяем и перемешиваем
    return pl.concat([pos, neg_sampled], how="vertical").sample(fraction=1.0, shuffle=True, seed=42)
```

```python
print("Создаем train и val датасеты...")
train = create_dataset_with_hard_negatives(hist_train, target_train, n_negatives=10)
val = create_dataset_with_hard_negatives(hist_val, target_val, n_negatives=10)
```

```python
# --- 4. ПОДСЧЕТ ФИЧЕЙ (твои функции) ---
def add_item_popularity(train_hist: pl.DataFrame, df: pl.DataFrame) -> pl.DataFrame:
    pop = train_hist.group_by("item_id").len().rename({"len": "item_popularity"})
    return df.join(pop, on="item_id", how="left").with_columns(pl.col("item_popularity").fill_null(0))

def add_user_count_likes(train_hist: pl.DataFrame, df: pl.DataFrame) -> pl.DataFrame:
    pop = train_hist.group_by("uid").len().rename({"len": "user_count_likes"})
    return df.join(pop, on="uid", how="left").with_columns(pl.col("user_count_likes").fill_null(0))

def add_item_organic_share(train_hist: pl.DataFrame, df: pl.DataFrame) -> pl.DataFrame:
    share = train_hist.group_by("item_id").agg(pl.col("is_organic").mean().alias("item_organic_share"))
    return df.join(share, on="item_id", how="left").with_columns(pl.col("item_organic_share").fill_null(0))

def add_user_organic_share(train_hist: pl.DataFrame, df: pl.DataFrame) -> pl.DataFrame:
    share = train_hist.group_by("uid").agg(pl.col("is_organic").mean().alias("user_organic_share"))
    return df.join(share, on="uid", how="left").with_columns(pl.col("user_organic_share").fill_null(0))

def add_user_artist_like_cnt(train_hist: pl.DataFrame, df: pl.DataFrame, item2artist: pl.DataFrame) -> pl.DataFrame:
    stats = (
        train_hist.select(["uid", "item_id"])
        .join(item2artist, on="item_id", how="left")
        .group_by(["uid", "artist_id"])
        .len()
        .rename({"len": "user_artist_like_cnt"})
    )
    return (
        df.join(item2artist, on="item_id", how="left")
          .join(stats, on=["uid", "artist_id"], how="left")
          .with_columns(pl.col("user_artist_like_cnt").fill_null(0))
          .drop('artist_id')
    )
```

```python
print("Считаем фичи для train...")
train = add_item_popularity(hist_train, train)
train = add_user_count_likes(hist_train, train)
train = add_item_organic_share(hist_train, train)
train = add_user_organic_share(hist_train, train)
train = add_user_artist_like_cnt(hist_train, train, artists)
```

```python
print("Считаем фичи для val...")
val = add_item_popularity(hist_val, val)
val = add_user_count_likes(hist_val, val)
val = add_item_organic_share(hist_val, val)
val = add_user_organic_share(hist_val, val)
val = add_user_artist_like_cnt(hist_val, val, artists)
```

```python
# --- 5. ОБУЧЕНИЕ МОДЕЛИ ---
train_pool = Pool(
    data=train.drop(['target', 'item_id', 'uid']).to_pandas(),
    label=train['target'].to_pandas(),
)

val_pool = Pool(
    data=val.drop(['target', 'item_id', 'uid']).to_pandas(),
    label=val['target'].to_pandas(),
)
```

```python
model = CatBoostClassifier(
    iterations=300, # Можно увеличить до 500, если не падает ОЗУ
    learning_rate=0.05, # Сделаем чуть поменьше для стабильности
    depth=5,
    l2_leaf_reg=5,
    loss_function="Logloss",
    eval_metric="AUC",
    early_stopping_rounds=50,
    verbose=50,
    use_best_model=True,
    task_type="GPU",
    train_dir='/tmp/catboost_info'# Если в Colab кончается GPU-память, поменяй на "CPU"
)

model.fit(train_pool, eval_set=val_pool, plot=False)
```

```python
# --- 6. ИНФЕРЕНС (ТЕСТ) ---
print("Генерация кандидатов для теста...")
# Берем топ-300 кандидатов из всей истории
test_candidates = get_recent_candidates(hist_test, top_n=300)
test = test_users.select(['uid']).join(test_candidates, how='cross')
```

```python
print("Считаем фичи для теста...")
test = add_item_popularity(hist_test, test)
test = add_user_count_likes(hist_test, test)
test = test.join(artists, on="item_id", how="left") # Добавляем артистов перед фичей
test = add_item_organic_share(hist_test, test)
test = add_user_organic_share(hist_test, test)
test = add_user_artist_like_cnt(hist_test, test, artists)
```

```python
test_pool = Pool(
    data=test.drop(['item_id', 'uid']).to_pandas(),
)
pred = model.predict_proba(test_pool)[:, 1]
```

```python
submit = (
    test.select(["uid", "item_id"])
    .with_columns(pl.Series("pred", pred))
    .sort(["uid", "pred"], descending=[False, True])
    .group_by("uid")
    .agg(
        # Обязательно берем 100 треков, как требует задача
        pl.col("item_id").head(100).cast(pl.Utf8).str.join(" ").alias("item_ids")
    )
)


submit.write_csv('catboost_improved.csv')
print("Готово!")
```
